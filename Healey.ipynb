# BASELINE METHOD

```python
import numpy as np
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
# from sklearn.metrics import precision_recall_fscore_support,confusion_matrix,precision_score,recall_score,accuracy_score
```

```python
PATH = "./predicting-driver-stress-using-deep-learning/scripts/csv_files/preprocessed_data/final_data/Drive05.csv"
df = pd.read_csv(PATH)
df["Stress_mean"] = df["Stress_mean"].replace([1],0)
df["Stress_mean"] = df["Stress_mean"].replace([3,5],1)
df = df.dropna()
df = df.reset_index()
df.tail(5)
```

```python
y = df["Stress_mean"]
X = df.drop(["time","Stress_mean"], axis=1)
print(y.value_counts())
```

```python
class_predict = []
class_true = []
for i in range(df.shape[0]):
    X_train = X.drop(index=[i])
    X_test = X.iloc[[i]]
    
    y_train = y.drop(index=[i])
    y_test = y.iloc[[i]]
    
#     lda = LinearDiscriminantAnalysis(n_components=2)
    lda = LinearDiscriminantAnalysis(n_components=1)
    X_train = lda.fit_transform(X_train, y_train)
#     X_test = lda.transform(X_test)
    y_predict = lda.predict(X_test)
    
    class_predict.append(y_predict[0])
    class_true.append(y_test)
```

```python
print(class_predict)
print(class_true)
```

```python
acc = accuracy_score(class_true, class_predict)
pred = precision_score(class_true, class_predict, average="macro")
recall = recall_score(class_true, class_predict, average="macro")
```

```python
print("Accuracy:{%f},Prediction:{%f},Recall:{%f}"%(acc,pred,recall))
```

# OUR METHOD

```python
PATH = "./predicting-driver-stress-using-deep-learning/scripts/csv_files/preprocessed_data/Drive05.csv"
df = pd.read_csv(PATH)
df["Stress"] = df["Stress"].replace([1],0)
df["Stress"] = df["Stress"].replace([3,5],1)
df = df.dropna()
df = df.reset_index()
df.head(5)
```

```python
df_EDA = df["hand GSR-mV"]
df_HR = df["HR-bpm"]
```

```python
import flirt
```

```python
new_df_EDA = flirt.eda.get_eda_features(data=df_EDA,window_length=10,window_step_size=10,data_frequency=31)
new_df_EDA
```

```python
df_stress = df["Stress"]
df_stress.index = pd.to_datetime(df_stress.index,unit='s')
df_stress_new = df_stress.resample('310S').apply(lambda x: x.mode()[0])
```

```python
df_stress = df_stress_new.reset_index(drop=True)
new_df_EDA.reset_index(drop=True,inplace=True)
```

```python
new_df_HR = pd.DataFrame()
df_HR.index = pd.to_datetime(df_HR.index, unit='s')
new_df_HR["HR_mean"] = df_HR.resample('310S').mean()
new_df_HR["HR_std"] = df_HR.resample('310S').std()
new_df_HR.reset_index(drop=True,inplace=True)
```

```python
df_our = pd.concat([new_df_EDA, new_df_HR,df_stress], axis=1)
index_HR = df_our[df_our["HR_mean"]<50].index
df_our.drop(index_HR , inplace=True)
df_our.reset_index(inplace=True)
```

```python
df_our.replace([np.inf, -np.inf], np.nan, inplace=True)
df_our = df_our.dropna(axis=1)
df_our.shape
data = df_our.drop(['Stress'], axis=1)
y = df_our["Stress"]
```

```python
scale = StandardScaler()
transformed_data = scale.fit_transform(data)
```

```python
pca = PCA(n_components=7)
principalComponents = pca.fit_transform(transformed_data)
n_pcs= pca.components_.shape[0]
```

```python
principalDf_columns = ["PC"+str(i+1) for i in range(n_pcs)]
principalDf = pd.DataFrame(data = principalComponents
             , columns = principalDf_columns)
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
principalDf.head(10)
```

```python
km = KMeans(n_clusters=2, init='k-means++', n_init=10, max_iter=300, tol=0.0001, verbose=0, random_state=None, copy_x=True, algorithm='lloyd')
```

```python
y_km = km.fit_predict(principalDf)
y_km
```

```python
accuracy_score(y_km,y)
```
